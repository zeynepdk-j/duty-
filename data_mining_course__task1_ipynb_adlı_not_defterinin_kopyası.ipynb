```python
 # bir haftalık program düzenleyicisi yapmak hedefim

def haftalik_plan_olustur():       #fonksiyonu tanımlıyoruz
    gunler = ["Pazartesi", "Salı", "Çarşamba", "Perşembe", "Cuma"]   #hafta içi günlerden oluşan bir liste oluşturuldu
    mesai_saatleri = ["09:00", "10:00", "11:00", "12:00", "13:00", "14:00", "15:00", "16:00"]  #ortalama mesai saatlerine göre 8 saatlik dilime bölündü gün

    haftalik_program = {}  # genel bir sözlük yapısı oluşturuldu

    print("---  HAFTALIK PROGRAM PLANLAYICI ---")    #kullanıcıya görünecek başlık program başladığına dair

    for gun in gunler:                            # for döngüsü ile günler listesindeki ("in" ile belirttik) her bir günü döndürecek
        print(f"\n<< {gun} günü planlanlarınızı giriyorsunuz:")   # "print" yazdırma fonksiyonu, kullanıcıya hangi gün planını girdiğini hatırlatıyor, "\n" boşluk için, "<<" belirteç olsun diye
        gunluk_gorevler = []                                     #bu planları gunluk_gorevler isimli listeye kaydedecek, geçici sonra birleştireceğiz

        for saat in mesai_saatleri:                           #mesai saatleri listesindeki her bir çalışma saatini döndürecek
            gorev = input(f"   Saat {saat} için görev: ")       # "input" girdiye imkan veren temel özelliklerden, her saat içindeki görevi girmesini istiyoruz
            if not gorev:                                       # "if not" ile gören girmezse, boş bırakılırsa belirtilmedi olarak kaydedecek
                            gorev = "Belirtilmedi"
            gunluk_gorevler.append(f"{saat} : {gorev}")         # .append birleştirme fonksiyonu-saat ve görevi birleştiriyor gunluk_gorevler adlı listeye

        haftalik_program [gun] = gunluk_gorevler         # ilgili günün tüm görev lstesini haftalik_program isimli sözlük yapısına atıyor


    with open("haftalik_program.txt", "w", encoding="utf-8") as dosya:             # bir dosya açıyoruz, "w" yazma komutu, utf-8 türkçe karakterleri algılasın, sonra bozulmasın diye
        dosya.write("=" * 40 + "\n")                                                # "dosya.write " yaz komutu, başlık üst ve altında iki çizgili basit bir görsel elde etmek için = işaretini 40 kez yazması ve alt satıra geçmesi "\n" ile
        dosya.write("       HAFTALIK ÇALIŞMA PROGRAMI        \n")                   # başlık ve alt satıra geçme
        dosya.write("========================================\n")                   # yine başlığın altını bait bir süsleme ve alt satıra geçme

        for gun, gorevler in haftalik_program.items():        # yine for döngüsüyle haftalık program içindeki tüm gün ve görevlerin her biri için aşağıdaki 4 satırı döndürecek
            dosya.write(f"\n<< {gun.upper()}\n")           # gün ismini büyük harfle yazacak
            dosya.write("-" * 20 + "\n")             #gün altına çizgi oluşacak
            for x in gorevler:                  # görevlerdeki her bir nesne için sırasıyla
                dosya.write(f"  [ ] {x}\n")       # bir kutucuk görevi yazıp aşağı satıra geçecek
            dosya.write("\n")                     #gün bitince bir satır boşluk verecek

    print("\n✅ Planınız başarıyla 'haftalik_plan.txt' dosyasına kaydedildi!") #işlem bittiğini haber verecek
```

```python
haftalik_plan_olustur()
```
